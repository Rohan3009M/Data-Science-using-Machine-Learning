# **Naive Bayes**
In this code I have used Naive Bayes Model to Predict whether Passenger on Titanic will survive or not.

```python
#Importing Required Libraries
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error, accuracy_score, f1_score
from sklearn.metrics import confusion_matrix, recall_score, classification_report
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

np.set_printoptions(precision=4, linewidth=100)

import warnings
warnings.filterwarnings('ignore')
```

```python
#Importing the Dataset
df = pd.read_csv("/content/Titanic.csv")
df.head()
```

```python
df.info()
```

```python
#Dropping non-usable features
df.drop(['PassengerId', 'Name', 'SibSp', 'Parch', 'Ticket', 'Cabin', 'Embarked'], axis = 'columns', inplace = True)
df.head()
```

```python
#Defining Dependent and Independent Variables i.e. 'Y' and 'x' resp.
y = df['Survived']
X = df.drop('Survived', axis = 'columns')
```

```python
dummies = pd.get_dummies(X.Sex)
dummies.head()
```

```python
X = pd.concat([X, dummies], axis = 'columns')
X.head()
```

```python
# Dropping sex column and male column
X.drop(['Sex', 'male'], axis = 'columns', inplace = True)
X.head()
```

```python
#Checking Missing Values
X.columns[X.isna().any()]
```

```python
X.Age[ : 10]
```

```python
#Changing Missing Values with mean of Age column
X.Age = X.Age.fillna(X.Age.mean())
X.head(10)
```

```python
#Now we split the data for training and testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
#Applying Model
model = GaussianNB()
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
X_test[0 : 10]
```

```python
y_test[0 : 10]
```

```python
#Making Prediciton
y_pred = model.predict(X_test)
y_pred
```

```python
model.predict_proba(X_test)
```

```python
#Checking Accuracy
cm = confusion_matrix(y_test, y_pred)
print (cm)
print(f1_score(y_test, y_pred))
```

```python
print(accuracy_score(y_test, y_pred))
```

```python
print( metrics.classification_report( y_test, y_pred ) )
```
