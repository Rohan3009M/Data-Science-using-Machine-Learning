# **Logistic Regression**
In this code I have used Logistic Regression Model to Predict whether to give Credit to a customer or not.

```python
#Importing Required Libraries
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error
from sklearn import metrics

import matplotlib.pyplot as plt
import seaborn as sn
%matplotlib inline

np.set_printoptions(precision=4, linewidth=100)

import warnings
warnings.filterwarnings('ignore')
```

```python
#Importing the Dataset
df = pd.read_csv( 'German Credit Data.csv' )
df.head( 10 )
```

```python
#Displaying features
df.iloc[0:5, 0:7]
```

```python
df.iloc[0:5, 7:]
```

```python
#Checking count of status for credit or not credit
df.status.value_counts()
```

```python
#Defining Independent Variables i.e. 'x'
X_features = list( df.columns )
X_features.remove( 'status' )
X_features
```

```python
#Encoding Categorical Features
encoded_df = pd.get_dummies( df[X_features], drop_first = True )
list(encoded_df.columns)
```

```python
#Defining Dependent Variables i.e. 'Y'
Y = df.status
X = sm.add_constant( encoded_df )
```

```python
#Now we split the data for training and testing
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.3, random_state = 42)
```

```python
#Applying Logit Model
logit = sm.Logit(y_train, X_train)
logit_model = logit.fit()
```

```python
logit_model.summary2()
```

```python
#Model Diagnosis
def get_significant_vars( lm ):
    var_p_vals_df = pd.DataFrame( lm.pvalues )
    var_p_vals_df['vars'] = var_p_vals_df.index
    var_p_vals_df.columns = ['pvals', 'vars']
    return list( var_p_vals_df[var_p_vals_df.pvals <= 0.05]['vars'] )
```

```python
significant_vars = get_significant_vars( logit_model )
significant_vars
```

```python
final_logit = sm.Logit( y_train, sm.add_constant( X_train[significant_vars] ) ).fit()
```

```python
final_logit.summary2()
```

```python
#Making Prediction
y_pred_df = pd.DataFrame( { "actual": y_test, "predicted_prob": final_logit.predict( sm.add_constant( X_test[significant_vars] ) ) } )   
y_pred_df.sample(10, random_state = 42)
```

```python
y_pred_df['predicted'] = y_pred_df.predicted_prob.map(lambda x: 1 if x > 0.5 else 0)
y_pred_df.sample(10, random_state = 42)
```

```python
#Creating Confusion Matrix
def draw_cm( actual, predicted ):
    cm = metrics.confusion_matrix( actual, predicted )
    sn.heatmap(cm, annot=True,  fmt='.2f', 
               xticklabels = ["Good credit", "Bad Credit"] , 
               yticklabels = ["Good credit", "Bad Credit"] )
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.show()
```

```python
draw_cm( y_pred_df.actual, 
        y_pred_df.predicted )
```

```python
#Measuring Accuracies

print( metrics.classification_report( y_pred_df.actual, y_pred_df.predicted ) )
```

```python
plt.figure( figsize = (8,6) )
sn.distplot( y_pred_df[y_pred_df.actual == 1]["predicted_prob"], 
            kde=False, color = 'b', 
            label = 'Bad Credit' )
sn.distplot( y_pred_df[y_pred_df.actual == 0]["predicted_prob"], 
            kde=False, color = 'g', 
            label = 'Good Credit' )
plt.legend()
plt.show()
```

```python
#ROC and AUC
def draw_roc( actual, probs ):
    fpr, \
    tpr, \
    thresholds = metrics.roc_curve( actual, probs, drop_intermediate = False )
    auc_score = metrics.roc_auc_score( actual, probs )
    plt.figure(figsize=(8, 6))
    plt.plot( fpr, tpr, label='ROC curve (area = %0.2f)' % auc_score )
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate or [1 - True Negative Rate]')
    plt.ylabel('True Positive Rate')
    plt.legend(loc="lower right")
    plt.show()
    
    return fpr, tpr, thresholds
```

```python
fpr, tpr, thresholds = draw_roc( y_pred_df.actual, y_pred_df.predicted_prob)
```
